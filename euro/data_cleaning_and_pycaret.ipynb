### KEEPING ONLY HIGH,LOW,MEDIUM IN ATTACK AND DEFENCE

```python
options = ['medium', 'high', 'low']
df1 = df[df.attacking_work_rate.isin(options)]
```

```python
df1
```

```python
options = ['medium', 'high', 'low']
df2 = df1[df1.defensive_work_rate.isin(options)]
```

```python
df2
```

### MAPPING STRINGS

```python
df3 = df2.copy()
```

```python
df3['preferred_foot']=df3['preferred_foot'].replace({'right':1,'left':2})
```

```python
df3['attacking_work_rate']=df3['attacking_work_rate'].replace({'low':0,'medium':1,'high':2})
```

```python
df3['defensive_work_rate']=df3['defensive_work_rate'].replace({'low':0,'medium':1,'high':2})
```
