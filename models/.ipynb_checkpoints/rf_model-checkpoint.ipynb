```python
from sklearn.ensemble import RandomForestClassifier
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
data = pd.read_csv('../data/ml_df.csv').drop(columns=['Unnamed: 0','away_team_api_id_HOME','home_team_api_id_HOME'])
```

```python
data
```

```python
listie =[]
for i in data['y']:
    if i == 1:
        stat = 'win'
    elif i == 0:
        stat = 'loss'
    else:
        stat = 'draw'
    listie.append(stat)

data['y'] = listie
        
```

```python
# Create a random forest classifier
data.fillna(data.mean(), inplace=True)

rf = RandomForestClassifier(n_estimators=200,random_state=33)
```

```python
data = data.drop('YEAR',axis=1)
```

```python
# Assign X (data) and y (target)
X = data.drop("y", axis=1)
y = data["y"]
print(X.shape, y.shape)
```

Split our data into training and testing

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
rf = rf.fit(X_train, y_train)
```

```python
rf.score(X_train, y_train)
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
importances
```

```python
# We can sort the features by their importance
sorted(zip(rf.feature_importances_, data.columns), reverse=True)
```
